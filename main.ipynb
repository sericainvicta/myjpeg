```python
import PIL
import numpy as np

from PIL import Image
from matplotlib import pyplot as plt
```

```python
im = Image.open('./data/ste.png')
print(im.format, im.size, im.mode)
plt.imshow(im)
```

# Step 1

 Image converted from the RGB color space to the YCbCr color space.

```python
im_rgb = im.convert('RGB')
im_yuv = im.convert('YCbCr')
pixels_rgb = np.asarray(im_rgb)
pixels_yuv = np.asarray(im_yuv)
pixels_rgb, pixels_yuv
```

```python
# 这里先取16x16块实验
im_yuv = im_yuv.crop((0, 0, 16, 16))
plt.imshow(im_yuv)
```

Sampled to YCbCr 4:2:0 format

# Step 2

splits the image into 8×8 blocks

# Step 3

 The Discrete Cosine
Transform (DCT) is performed on each block to transform
from the pixel domain to the frequency domain

# Step 4

quantization
operation is used to simplify the transformed components
with information reduction. This is done by using a predefined quantization table to round the DCT coefficients based
on the human sensitivity to different frequencies.

# Step 5

JPEG encodes the quantized data using Huffman coding [11].
Huffman coding is an entropy coding technique that assigns
different lengths to the data based on how often it appears.
